# Visualization

### Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
aapl = pd.read_csv("AAPL.csv", index_col="Date", parse_dates=True)
```

```python
aapl.head()
```

```python
fig, ax = plt.subplots()
aapl['Close'].plot(ax=ax)
```

```python
aapl['MA5'] = aapl['Close'].rolling(5).mean()
aapl['MA20'] = aapl['Close'].rolling(20).mean()
```

```python
fig, ax = plt.subplots()
aapl[['MA5', 'MA20', 'Close']].plot(ax=ax)
```

```python
fig, ax = plt.subplots()
aapl[['MA5', 'MA20', 'Close']].loc['2020':].plot(ax=ax)
```
