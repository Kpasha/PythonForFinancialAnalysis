# Correlation

```python
import pandas_datareader as pdr
import datetime as dt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
tickers = ['SPY', 'TLT']

start = dt.datetime(1999, 1, 1)

data = pdr.get_data_yahoo(tickers, start)
```

```python
data.dropna()['Adj Close']
```

```python
data = data.dropna()['Adj Close']
```

```python
log_returns = np.log(data/data.shift())
```

```python
log_returns.corr()
```

```python
fig, ax = plt.subplots()
(data/data.iloc[0]).plot(ax=ax)
```

```python
data_set = data.loc['2008-05':'2011-04']

fig, ax = plt.subplots()
(data_set/data_set.iloc[0]).plot(ax=ax)
```
