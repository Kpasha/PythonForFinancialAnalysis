## Performance

### GAGR

```python
data = sp500['Close'].loc['1999':'2008']
```

```python
total_return = data.iloc[-1]/data.iloc[0]
```

```python
total_return
```

```python
(data.iloc[-1]/data.iloc[0])**(1/10) - 1
```

### Drawdown

```python
rolling_max = data.cummax()
daily_drawdown = data/rolling_max - 1
```

```python
max_drawdown = daily_drawdown.cummin().iloc[-1]
```

```python
max_drawdown
```
